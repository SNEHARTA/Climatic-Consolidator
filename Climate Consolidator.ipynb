```python
import requests
from collections import Counter
import numpy as np

# API keys for the respective weather services
api_keys = {
    "weatherapi": "a84f091267524e269ad140955241610",
    "openweathermap": "6596395ee420147e31f115949bfa46d7",
    "weatherstack": "29764a6dec199642c1f239121e1aa67d"
}

# Function to calculate the most frequent or closest value for numerical data
def get_closest_or_frequent(values):
    # If values are very close to each other, return the average
    if max(values) - min(values) < 2:  # Threshold for similarity, can be adjusted
        return sum(values) / len(values)
    else:
        # Find the value closest to the median
        median_value = np.median(values)
        closest_value = min(values, key=lambda x: abs(x - median_value))
        return closest_value

# Function to get the most frequent value for categorical data
def get_most_frequent(values):
    return Counter(values).most_common(1)[0][0]

def get_weather_data_from_weatherapi(location):
    try:
        complete_api_link = f"https://api.weatherapi.com/v1/current.json?key={api_keys['weatherapi']}&q={location}"
        api_link = requests.get(complete_api_link)
        api_data = api_link.json()

        if 'error' in api_data:
            return None
        else:
            current_weather = api_data['current']
            return {
                "temperature": current_weather['temp_c'],
                "weather_desc": current_weather['condition']['text'],
                "humidity": current_weather['humidity'],
                "wind_speed": current_weather['wind_kph'],
                "clouds": current_weather['cloud'],
                "visibility": current_weather['vis_km']
            }
    except Exception as e:
        return None

def get_weather_data_from_openweathermap(location):
    try:
        complete_api_link = f"https://api.openweathermap.org/data/2.5/weather?q={location}&appid={api_keys['openweathermap']}"
        api_link = requests.get(complete_api_link)
        api_data = api_link.json()

        if api_data['cod'] == '404':
            return None
        else:
            temp_city = api_data['main']['temp'] - 273.15  # Convert Kelvin to Celsius
            return {
                "temperature": temp_city,
                "weather_desc": api_data['weather'][0]['description'],
                "humidity": api_data['main']['humidity'],
                "wind_speed": api_data['wind']['speed'],
                "clouds": api_data['clouds']['all'],
                "visibility": api_data['visibility'] / 1000  # Convert to kilometers
            }
    except Exception as e:
        return None

def get_weather_data_from_weatherstack(location):
    try:
        complete_api_link = f"https://api.weatherstack.com/current?access_key={api_keys['weatherstack']}&query={location}"
        api_link = requests.get(complete_api_link)
        api_data = api_link.json()

        if 'error' in api_data:
            return None
        else:
            current_weather = api_data['current']
            return {
                "temperature": current_weather['temperature'],
                "weather_desc": current_weather['weather_descriptions'][0],
                "humidity": current_weather['humidity'],
                "wind_speed": current_weather['wind_speed'],
                "clouds": current_weather['cloudcover'],
                "visibility": current_weather['visibility']
            }
    except Exception as e:
        return None

# Consolidate the weather data from all sources
def consolidate_weather_data(location):
    weather_data_sources = [
        get_weather_data_from_weatherapi(location),
        get_weather_data_from_openweathermap(location),
        get_weather_data_from_weatherstack(location)
    ]

    # Filter out None results
    weather_data_sources = [data for data in weather_data_sources if data]

    if not weather_data_sources:
        return None

    def consolidate_value(key):
        values = [data[key] for data in weather_data_sources if data[key] is not None]
        if isinstance(values[0], str):
            return get_most_frequent(values)  # For categorical values like weather_desc
        else:
            return get_closest_or_frequent(values)  # For numerical values like temperature

    consolidated_data = {
        "temperature": consolidate_value("temperature"),
        "weather_desc": consolidate_value("weather_desc"),
        "humidity": consolidate_value("humidity"),
        "wind_speed": consolidate_value("wind_speed"),
        "clouds": consolidate_value("clouds"),
        "visibility": consolidate_value("visibility")
    }

    return consolidated_data

def main():
    while True:
        location = input("Enter the city name (or type 'quit' to exit): ")
        if location.lower() == 'quit':
            break
        weather_data = consolidate_weather_data(location)
        if weather_data is None:
            print(f"Invalid City: {location}, Please check your City name")
        else:
            print("----------------------------------------------------------------")
            print(f"Weather Stats for {location.upper()}")
            print("-----------------------------------------------------------------")
            print(f"Temperature: {weather_data['temperature']}째C")
            print(f"Weather Description: {weather_data['weather_desc']}")
            print(f"Humidity: {weather_data['humidity']}%")
            print(f"Wind Speed: {weather_data['wind_speed']} kph")
            print(f"Clouds: {weather_data['clouds']}%")
            print(f"Visibility: {weather_data['visibility']} Km")
            print("----------------------------------------------------------------")

if __name__ == "__main__":
    main()
```
